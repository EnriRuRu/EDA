Ahora tengo que sacar dos datas de data1, para hazarous TRUE y para False

```python
#del data principal los que no se repiten, los TRUE
data1_ht= data1[data1["hazardous"]==True]
data1_ht
```

```python
#del data principal los que no se repiten, los FALSE
data1_hf= data1[data1["hazardous"]==False]
data1_hf
```

VELOCIDADES

```python
# DATA FINAL VELOCIDAD NO AMENAZANTE
# data1_hf con cosa0_vel (no amenazantes (H=FALSE) con los repetidos con velocidad máxima)
data_false_vel= pd.concat([data1_hf, cosa0_vel])
```

```python
# DATA FINAL VELOCIDAD SÏ AMENAZANTE
# data1_ht con cosa0_vel (sí amenazantes (H=TRUE) con los repetidos con velocidad máxima)

data_true_vel= pd.concat([data1_ht, cosa1_vel])
```

DISTANCIAS

```python
# DATA FINAL DISTANCIA NO AMENAZANTE
# data1_hf con cosa0_miss_distance (no amenazantes (H=FALSE) con los repetidos con distancia mínima)
data_false_dist= pd.concat([data1_hf, cosa0_miss_distance])
```

```python
# DATA FINAL DISTANCIA SÍ AMENAZANTE
# data1_hT con cosa0_miss_distance (sí amenazantes (H=TRUE) con los repetidos con DISTANCIA MÍNIMA)
data_true_dist= pd.concat([data1_ht, cosa1_miss_distance])
```

DIÁMETRO

```python
# DATA FINAL DIÁMETRO NO AMENAZANTE
# data1_hf con cosa0_diam (no amenazantes (H=FALSE) con los repetidos con diámetro máxima)
data_false_diam= pd.concat([data1_hf, cosa0_diam])
```

```python
# DATA FINAL DIÁMETRO SÎ AMENAZANTE
# data1_ht con cosa1_diam (sí amenazantes (H=TRUE) con los repetidos con diámetro máxima)
data_true_diam= pd.concat([data1_ht, cosa1_diam])
data_true_diam

```
